## Web-scraping - Game of Thrones - Episode Scripts Of All Seasons

**Quick Note**

This Notebook gets all starring cast characters of the popular HBO series 'Game of Thrones'.

### Importing the necessary tools

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Source

```python
page_url = 'https://en.wikipedia.org/wiki/List_of_Game_of_Thrones_characters'
```

```python
# Don't need to verify certificate for wikipedia...
import ssl
ssl._create_default_https_context = ssl._create_unverified_context
```

### Fetch tables from url

```python
table_MN = pd.read_html(page_url)
len(table_MN)
```

### Identify tables of interest

#### Main characters

```python
main_cast = table_MN[3]
main_cast.head()
```

```python
main_cast.shape
```

```python
main_char = main_cast[['Character', 'Appearances']]

main_char.head()
```

#### Recurring characters

```python
recurr_cast = table_MN[4]
recurr_cast.head()
```

```python
recurr_cast.shape
```

```python
recurr_char = recurr_cast[['Character', 'Appearances']]

recurr_char.head()
```

#### Merge 2 character tables

```python
characters = pd.concat([main_char, recurr_char], ignore_index=True)

characters.shape

characters
```

### Create 'characters.csv'

```python
string = characters.to_csv(sep=',')

f = open('characters.csv',"w+")
f.write(string)
```
