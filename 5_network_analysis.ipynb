### Import the necessary tools

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import os
```

## 1 - Load Graphs

```python
# All season directories
all_graphs = [g for g in os.scandir('Graphs_gml') if '.gml' in g.name]

# Sort directories
all_graphs.sort(key=lambda x: x.name)
```

```python
season_graphs = []
# Get graphs and append to list
for gml in all_graphs:
    file = str(gml.name)
    graph = nx.read_gml('Graphs_gml/'+file)
    season_graphs.append(graph)
```

```python
season_graphs
```

### 4.1 - Centrality measures

#### 4.1.1 - Degree centrality

```python
counter = 0

for g in season_graphs:
    degree_dict = nx.degree_centrality(g)
    degree_df = pd.DataFrame.from_dict(degree_dict, orient='index', columns=['centrality'])
    cur = degree_df.sort_values('centrality', ascending=False)[0:9]

    fig = plt
    fig = cur.plot(kind='bar')
    fig.set_title('Season '+str(counter + 1))
    counter += 1
```

```python
house_stark = ['Eddard Stark', 'Catelyn Stark','Robb Stark', 'Arya Stark', 'Sansa Stark', 'Jon Snow', 'Bran Stark', 'Rickon Stark']

index = ['s1', 's2', 's3', 's4','s5', 's6', 's7', 's8']

evol = [nx.degree_centrality(season) for season in season_graphs]

degree_evol_df = pd.DataFrame.from_records(evol)

degree_evol_df[house_stark].plot(figsize=(15,8))
```
